# Data Cleaning

# 📈 Key Performance Indicators (KPIs)

```python
# Total Sales
total_sales = df["Sales"].sum()

print(f"Total Sales: ${total_sales:,.2f}")
```

```python
# Total Profit
total_profit = df["Profit"].sum()

print(f"Total Profit: ${total_profit:,.2f}")
```

```python
# Total Orders
total_orders = df["Order ID"].nunique()

print(f"Total Orders: {total_orders}")
```

```python
# Average Sales per Order
average_sales = df["Sales"].mean()

print(f"Average Sales: ${average_sales:,.2f}")
```

```python
# Average Profit per Order
average_profit = df["Profit"].mean()

print(f"Average Profit: ${average_profit:,.2f}")
```

# 📊 Sales Analysis

```python
# Sales by Category
sales_by_category = df.groupby("Category")["Sales"].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))

sales_by_category.plot(kind="bar")

plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.xticks(rotation=0)
plt.tight_layout()

plt.show()
```

```python
# Sales by Sub-Category
sales_by_subcategory = (
    df.groupby("Sub-Category")["Sales"]
    .sum()
    .sort_values(ascending=False)
)

plt.figure(figsize=(12,6))

sales_by_subcategory.plot(kind="bar")

plt.title("Sales by Sub-Category")
plt.xlabel("Sub-Category")
plt.ylabel("Total Sales")
plt.xticks(rotation=60, ha="right")
plt.tight_layout()

plt.show()
```

```python
# Sales by Region
sales_by_region = (
    df.groupby("Region")["Sales"]
    .sum()
    .sort_values(ascending=False)
)

plt.figure(figsize=(8,5))

sales_by_region.plot(kind="bar")

plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.xticks(rotation=0)
plt.tight_layout()

plt.show()
```

```python
# Sales by Segment
sales_by_segment = (
    df.groupby("Segment")["Sales"]
    .sum()
    .sort_values(ascending=False)
)

plt.figure(figsize=(8,5))

sales_by_segment.plot(kind="bar")

plt.title("Sales by Segment")
plt.xlabel("Segment")
plt.ylabel("Total Sales")
plt.xticks(rotation=0)
plt.tight_layout()

plt.show()
```

# 💰 Profit Analysis

```python
# Monthly Profit Trend
monthly_profit = (
    df.groupby(df["Order Date"].dt.to_period("M"))["Profit"]
      .sum()
)

monthly_profit.index = monthly_profit.index.astype(str)

plt.figure(figsize=(12,5))
monthly_profit.plot()

plt.title("Monthly Profit Trend")
plt.xlabel("Month")
plt.ylabel("Profit")
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
print(df["Order Date"].dtype)
```

```python
# Profit by Category
profit_by_category = (
    df.groupby("Category")["Profit"]
    .sum()
    .sort_values(ascending=False)
)

plt.figure(figsize=(8,5))

profit_by_category.plot(kind="bar")

plt.title("Profit by Category")
plt.xlabel("Category")
plt.ylabel("Total Profit")
plt.xticks(rotation=0)
plt.tight_layout()

plt.show()
```

```python
# Discount vs Profit
plt.figure(figsize=(8,5))

plt.scatter(
    df["Discount"],
    df["Profit"],
    alpha=0.5
)

plt.title("Discount vs Profit")
plt.xlabel("Discount")
plt.ylabel("Profit")
plt.tight_layout()

plt.show()
```

```python
# Profit by Region
profit_by_region = (
    df.groupby("Region")["Profit"]
    .sum()
    .sort_values(ascending=False)
)

plt.figure(figsize=(8,5))

profit_by_region.plot(kind="bar")

plt.title("Profit by Region")
plt.xlabel("Region")
plt.ylabel("Total Profit")
plt.xticks(rotation=0)
plt.tight_layout()

plt.show()
```

# 👥 Customer Analysis

```python
# Top 10 Customers by Sales
top_customers = (
    df.groupby("Customer Name")["Sales"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(10,6))

top_customers.plot(kind="barh")

plt.title("Top 10 Customers by Sales")
plt.xlabel("Total Sales")
plt.ylabel("Customer")
plt.tight_layout()

plt.show()
```

# 📦 Product Analysis

```python
# Top 10 Products by Sales
top_products = (
    df.groupby("Product Name")["Sales"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(12,6))

top_products.plot(kind="barh")

plt.title("Top 10 Products by Sales")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.xticks(rotation=60, ha="right")
plt.tight_layout()

plt.show()
```

```python
# Top 10 Loss-Making Products
loss_products = (
    df.groupby("Product Name")["Profit"]
      .sum()
      .sort_values(ascending=True)
      .head(10)
)

plt.figure(figsize=(12,7))

loss_products.plot(kind="barh")

plt.title("Top 10 Loss-Making Products", fontsize=14, fontweight="bold")
plt.xlabel("Total Loss")
plt.ylabel("Product")

plt.grid(axis="x", linestyle="--", alpha=0.5)

plt.tight_layout()

plt.show()
```

# 🚚 Shipping Analysis

```python
# Sales by Ship Mode
shipping_sales = (
    df.groupby("Ship Mode")["Sales"]
    .sum()
    .sort_values(ascending=False)
)

plt.figure(figsize=(8,5))

shipping_sales.plot(kind="bar")

plt.title("Sales by Ship Mode")
plt.xlabel("Ship Mode")
plt.ylabel("Total Sales")
plt.xticks(rotation=0)
plt.tight_layout()

plt.show()
```

# 💡 Business Insights

## Key Business Insights

- Technology generated the highest overall sales and profit, making it the strongest-performing category.

- Furniture contributed significant sales but showed comparatively lower profitability, indicating potential pricing or cost challenges.

- A few products consistently generated losses, suggesting they should be reviewed for pricing, discounting, or discontinuation.

- Higher discounts were generally associated with lower profits, highlighting the impact of excessive discounting on business performance.

- The West region achieved the highest profitability, while other regions may require strategic improvements.

- Standard Class handled the largest share of sales, making it the most preferred shipping mode.

- Sales and profits varied across months, indicating seasonal demand patterns that can support inventory and marketing planning.
